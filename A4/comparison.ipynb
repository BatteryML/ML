# time_effect_plot

```python
with open('./pi_forest_result.p', 'rb') as f:
    iter_pi_forest_small, iter_pi_forest_large, t_pi_forest_small, t_pi_forest_large = pickle.load(f)
with open('./vi_forest_result.p', 'rb') as f:
    iter_vi_forest_small, iter_vi_forest_large, t_vi_forest_small, t_vi_forest_large = pickle.load(f)
with open('./ql_forest_result.p', 'rb') as f:
    iter_ql_forest_small, iter_ql_forest_large, t_ql_forest_small, t_ql_forest_large = pickle.load(f)
with open('./pi_lake_result.p', 'rb') as f:
    iter_pi_lake_small, iter_pi_lake_large, t_pi_lake_small, t_pi_lake_large = pickle.load(f)
with open('./vi_lake_result.p', 'rb') as f:
    iter_vi_lake_small, iter_vi_lake_large, t_vi_lake_small, t_vi_lake_large = pickle.load(f)
with open('./ql_lake_result.p', 'rb') as f:
    iter_ql_lake_small, iter_ql_lake_large, t_ql_lake_small, t_ql_lake_large = pickle.load(f)
```

```python
iter_list = [
    [iter_pi_forest_small+1, iter_pi_forest_large+1],
    [iter_vi_forest_small, iter_vi_forest_large],
    [iter_ql_forest_small, iter_ql_forest_large],
]
time_list = [
    [round(t_pi_forest_small, 4), round(t_pi_forest_large, 4)],
    [round(t_vi_forest_small, 4), round(t_vi_forest_large, 4)],
    [round(t_ql_forest_small, 4), round(t_ql_forest_large, 4)],
]
all_time_effect_plot(iter_list, time_list, labels=['PI', 'VI', 'Q-Learning'], 
                     problem='Forest', logy=[True, True])
```

```python
iter_list = [
    [iter_pi_lake_small+1, iter_pi_lake_large+1],
    [iter_vi_lake_small, iter_vi_lake_large],
    [iter_ql_lake_small, iter_ql_lake_large],
]
time_list = [
    [round(t_pi_lake_small, 4), round(t_pi_lake_large, 4)],
    [round(t_vi_lake_small, 4), round(t_vi_lake_large, 4)],
    [round(t_ql_lake_small, 4), round(t_ql_lake_large, 4)],
]
all_time_effect_plot(iter_list, time_list, labels=['PI', 'VI', 'Q-Learning'], 
                     problem='Lake', logy=[True, True])
```
